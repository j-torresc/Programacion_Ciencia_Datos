Se consolidan los Dataframes inicialmente cargados ("Orders" y "Details") utilizando el metodo merge por la variable "Order ID" (común en ambos Dataframes)

```python
#unir dataframes por la columna Order ID
df_final = pd.merge(df,df_1, on='Order ID', how='inner')

#visualiza primeros registros 
print('Primeros registros del Dataframe unificado:')
df_final.head()
```

Luego de consolidar los dataframe 'Orders' y 'Details' se logra visualizar el conjunto de datos con las columnas e información de ambos dataframes.

```python
#total ventas por categoria
ventas_categoria = df_final.groupby('Category')['Amount'].sum()

print('Ventas por Categoria:')
print(ventas_categoria)
```

## Visualizaciones

```python
#crea grafico de barras
plt.figure(figsize=(8, 6))
ventas_categoria = df_final.groupby('Category')['Amount'].sum().reset_index()
sns.barplot(x='Category', y='Amount', data=ventas_categoria)
plt.title('Ventas Totales por Categoria')
plt.xlabel('Categoria')
plt.ylabel('Ventas Totales')
plt.show()
```

Validamos que la categoria que mas ventas($) genera es la de electronicos, seguido por ropa y por último muebles.

```python
# Crear gráfico de pastel
plt.figure(figsize=(8, 8))
plt.pie(df_final['PaymentMode'].value_counts(), 
        labels=df_final['PaymentMode'].value_counts().index, autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title('Participación Porcentual de Cada Metodo de pago')
plt.show()
```

Se puede concluir que el método de pago mas utilizado es COD. 

```python
#Grafico dispersion
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Amount', y='Profit', data=df_final)
plt.title('Monto vs Beneficio')
plt.xlabel('Monto')
plt.ylabel('Beneficio')
plt.show()
```

```python
#crea grafico de barras
plt.figure(figsize=(8, 6))
cantidad_productos_categoria = df_final.groupby('Category')['Quantity'].sum().reset_index()
sns.barplot(x='Category', y='Quantity', data=cantidad_productos_categoria)
plt.title('Número de Pedidos por Categoría')
plt.xlabel('Categoría')
plt.ylabel('Número de Pedidos')
plt.show()
```

La categoria que mas cantidad de productos vende es ropa, seguido de electronicos y por ultimo muebles.

```python
#Grafico de cajas
plt.figure(figsize=(10, 6))
sns.boxplot(x='Category', y='Profit', data=df_final)
plt.title('Beneficio por Categoría')
plt.xlabel('Categoría')
plt.ylabel('Beneficio')
plt.show()
```

```python
# obtiene el monto de ventas por estado y categoría
ventas_por_estado_categoria = df_final.groupby(['State', 'Category'])['Amount'].sum().unstack()

# Crear el gráfico de barras apiladas
ventas_por_estado_categoria.plot(kind='bar', stacked=True, figsize=(12, 8), colormap='tab20')

plt.title('Monto de Ventas por Estado y Categoría')
plt.xlabel('Estado')
plt.ylabel('Monto de Ventas')
plt.legend(title='Categoría')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Se identifica que el estado que mas ventas($) genera es Maharashtra seguido de Madhya Pradesh. Así mismo, se muestra la distribución de las categorias vendidas por cada estado.

Se puede concluir que:

*  La categoria que mas ventas genera es Electronicos.
*  El metodo de pago mas utilizado es COD.
*  La categoria que mas cantidad de productos vende es ropa.
*  El estado donde se genera mayores ventas($) es Maharashtra
